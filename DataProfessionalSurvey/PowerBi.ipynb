```python
import pandas as pd
```

```python
excel_file = pd.ExcelFile('powerBIproject.xlsx')
```

```python
print(excel_file.sheet_names) 
```

```python
df = excel_file.parse('Data Professional Survey') 
print(df)
```

```python
#remove 'Other (Please Specify):' so we can only see what comes after :
import pandas as pd

df_copy = df.copy()

# Assuming 'YourColumn' is the column containing the strings
df_copy['Q1 - Which Title Best Fits your Current Role?'] = df_copy['Q1 - Which Title Best Fits your Current Role?'].str.replace(r'.*:', '', regex=True)   
df_copy['Q4 - What Industry do you work in?'] = df_copy['Q4 - What Industry do you work in?'].str.replace(r'.*:', '', regex=True)
df_copy['Q8 - If you were to look for a new job today, what would be the most important thing to you?'] = df_copy['Q8 - If you were to look for a new job today, what would be the most important thing to you?'].str.replace(r'.*:', '', regex=True)
df_copy['Q11 - Which Country do you live in?'] = df_copy['Q11 - Which Country do you live in?'].str.replace(r'.*:', '', regex=True)
df_copy['Q13 - Ethnicity'] = df_copy['Q13 - Ethnicity'].str.replace(r'.*:', '', regex=True)

# Display the DataFrame
print(df_copy.head(10))
```

```python
#remove Other (Please Specify)
# Assuming df is your DataFrame
contains_other = df_copy.applymap(lambda x: 'Other (Please Specify)' in str(x))

# Filter rows where the specified string is present in any column
rows_with_other = df_copy[contains_other.any(axis=1)]

# Display the rows where the specified string is present
print(rows_with_other)

```

```python
#replace Other (Please Specify) with Other
# Assuming df is your DataFrame
columns_to_replace = ['Q11 - Which Country do you live in?', 'Q13 - Ethnicity']

# Replace 'Other (Please Specify)' with 'Other' in specified columns
df_copy[columns_to_replace] = df_copy[columns_to_replace].replace('Other (Please Specify)', 'Other', regex=True)
```

```python
# Assuming df is your DataFrame
contains_other = df_copy.applymap(lambda x: 'Other(Please Specify)' in str(x))

# Filter rows where the specified string is present in any column
rows_with_other = df_copy[contains_other.any(axis=1)]

# Display the rows where the specified string is present
print(rows_with_other)
```

```python
# Assuming df is your DataFrame
print(df_copy['Q3 - Current Yearly Salary (in USD)'].head())
```

```python
#calculate averages from salary ranges given per row
# Assuming df is your DataFrame
df_copy['Q3 - Current Yearly Salary (in USD)'] = df_copy['Q3 - Current Yearly Salary (in USD)'].apply(lambda x: x.replace('k', '').replace('+', ''))

# Split the salary range and convert to integers
df_copy['Q3 - Current Yearly Salary (in USD)'] = df_copy['Q3 - Current Yearly Salary (in USD)'].apply(lambda x: [int(s) for s in x.split('-')])

# Calculate the average salary
df_copy['Q3 - Current Yearly Salary (in USD)'] = df_copy['Q3 - Current Yearly Salary (in USD)'].apply(lambda x: sum(x) / len(x) if len(x) > 0 else 0)

# Add 'k' back
df_copy['Q3 - Current Yearly Salary (in USD)'] = df_copy['Q3 - Current Yearly Salary (in USD)'].astype(int).astype(str) + 'k'

# Show the modified column
print(df_copy['Q3 - Current Yearly Salary (in USD)'].head(10))

```

```python
print(df_copy['Q3 - Current Yearly Salary (in USD)'].head())
```

```python
def extract_salary(value):
    if 'k' in value:
        return int(value.replace('k', ''))
    else:
        return int(value)

# Apply the function to create the 'Average Salary' column
df_copy['Average Salary'] = df_copy['Q3 - Current Yearly Salary (in USD)'].apply(extract_salary)

# Display the updated DataFrame
print(df_copy['Average Salary'].head())
```
